```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv',encoding= 'unicode_escape')
df
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.drop(['Status','unnamed1'],axis = 1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
pd.isnull(df).sum()
```

```python
df.rename(columns = {'Marital_Status': 'Married'})
df
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

# which gender is buying most

```python
gc = sns.countplot(x='Gender', data = df)
for bars in gc.containers:
    gc.bar_label(bars)
```

```python
sale_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data = sale_gender)
```

# Age

```python
ac = sns.countplot(x = 'Age Group' , data = df, hue = 'Gender')
for bars in ac.containers:
    ac.bar_label(bars)
```

```python
sale_ac = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.barplot(x='Age Group', y = 'Amount', data = sale_ac)
```

# State

```python
sale_sc = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State', y = 'Orders', data = sale_sc)
```

```python
sale_sc = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State', y = 'Amount', data = sale_sc)
```

# Marital_Status

```python
ms = sns.countplot(x = 'Marital_Status', data = df)
sns.set(rc={'figure.figsize':(10,3)})
for bars in ms.containers:
    ms.bar_label(bars)
```

```python
sale_ms = df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending=False)
sns.set(rc={'figure.figsize':(7,4)})
sns.barplot(x= 'Marital_Status', y = "Amount", data = sale_ms, hue ='Gender')
```

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
oc = sns.countplot(x= "Occupation", data = df)
for bars in oc.containers:
    oc.bar_label(bars)
```

```python
sale_oc = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x= 'Occupation', y = 'Amount', data = sale_oc)
```

# Product_Category

```python
pc = sns.countplot(x='Product_Category', data = df)
sns.set(rc={'figure.figsize':(25,7)})
for bars in pc.containers:
    pc.bar_label(bars)
```

```python
sale_pc = df.groupby(["Product_Category"], as_index=False)['Amount'].sum().sort_values(by = 'Amount', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,3)})
sns.barplot(x= 'Product_Category', y = 'Amount', data =sale_pc)
```

# Product_ID

```python
sale_pc = df.groupby(["Product_ID"], as_index=False)['Orders'].sum().sort_values(by = 'Orders', ascending= False).head(10)
sns.set(rc={'figure.figsize':(20,3)})
sns.barplot(x= 'Product_ID', y = 'Orders', data =sale_pc)
```
